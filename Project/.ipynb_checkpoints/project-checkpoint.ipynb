### Remove Unwanted Columns

```python
df.drop(['policy_id'],axis=1,inplace=True)
```

### Convert Object into numerical values

```python
df["population_density"] = pd.to_numeric(df["population_density"],errors='coerce')
df.head()
```

```python
df=pd.get_dummies(df,columns=['segment','model','fuel_type','engine_type','rear_brakes_type','transmission_type','steering_type'],dtype=int)
```

### Converting Categorical columns into numerical

```python

from sklearn.preprocessing import OrdinalEncoder
cols = [    'is_esc',
    'is_adjustable_steering',
    'is_tpms',
    'is_parking_sensors',
    'is_parking_camera',
    'is_front_fog_lights',
    'is_rear_window_wiper',
    'is_rear_window_washer',
    'is_rear_window_defogger',
    'is_brake_assist',
    'is_power_door_locks',
    'is_central_locking',
    'is_power_steering',
    'is_driver_seat_height_adjustable',
    'is_day_night_rear_view_mirror',
    'is_ecw',
    'is_speed_alert',
    'area_cluster'
    ]
df[cols] = OrdinalEncoder().fit_transform(df[cols])
```

```python
df.head()
```

### Split columns

```python
# Splitting max_power column into power & power_rpm
df[['power','power_rpm']]=df['max_power'].str.extract(r'(\d+)\D+(\d+)')
df[['power','power_rpm']]=df[['power','power_rpm']].astype(float)
```

```python
# Splitting max_torque column into torque & torque_rpm
df[['torque','rpm']]=df['max_torque'].str.extract(r'(\d+)\D+(\d+)')
df[['torque','rpm']]=df[['torque','rpm']].astype(float)
```

```python
# extracting numerical values from Length and Width
df['Length'] = df['Length'].str.extract(r'(\d+)').astype(float)
df['width']  = df['width'].str.extract(r'(\d+)').astype(float)
```

```python
#extracting numerical values from Gross_weight
df['Gross_weight']  = df['Gross_weight'].str.extract(r'(\d+)').astype(float)
```

```python
df.drop(['max_torque','max_power'],axis=1,inplace = True)
```

```python
df['torque/rpm'] = df['torque']/df['rpm']
df['power/rpm'] = df['power']/df['power_rpm']
```

### Scalling

### Split Data

```python
x = df.drop(['make','is_claim'],axis=1)
y = df['is_claim']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

### Smote

```python
from imblearn.combine import SMOTETomek
smote_tomek = SMOTETomek(random_state=42)
x_train_sm, y_train_sm = smote_tomek.fit_resample(x_train, y_train)
```

### Functions Of All Models

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegressionCV

def logistic_model():
    model = LogisticRegressionCV(cv=5,max_iter=1000)
    model.fit(x_train_sm,y_train_sm)
    return model
```

#### KNN

```python
from  sklearn.neighbors import KNeighborsClassifier

def knn_model() :
    model = KNeighborsClassifier(n_neighbors = 100)
    model.fit(x_train_sm,y_train_sm)
    return model
```

#### SVM

```python
from sklearn.svm import SVC

def svm_model():
    model = SVC()
    model.fit(x_train_sm,y_train_sm)
    return model
```

#### Decision Tree

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
def decision_model():
    model = DecisionTreeClassifier()

    param = {
    "criterion" : ["gini","log_loss","entropy"],
    "max_depth" : [10,20,30,40,50,60]
    }

    grid_tree = GridSearchCV(model,param)
    grid_tree.fit(x_train_sm,y_train_sm)
    return grid_tree.best_estimator_
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def evaluate_model(model):
    y_pred = model.predict(x_test)
    testing_accuracy = accuracy_score(y_test, y_pred)
    testing_precision = precision_score(y_test, y_pred,average='weighted')
    testing_recall = recall_score(y_test, y_pred,average='weighted')
    testing_f1 = f1_score(y_test, y_pred,average='weighted')

    return testing_accuracy, testing_precision, testing_recall, testing_f1
```

### Generate Performance Chart

```python
models = [
    ("Logistic Regression",logistic_model(),(0,0)),
    ("KNN",knn_model(),(0,1)),
    ("SVM",svm_model(),(1,0)), 
    ("Decision Tree",decision_model(),(1,1))
]
```

```python
performance_data = []
for model_name,model,position in models :
    test_accuracy,precision,recall,f1 = evaluate_model(model)
    performance_data.append([
        model_name,f"{test_accuracy*100:0.2f}%",f"{precision*100:0.2f}%",f"{recall*100:0.2f}%",f"{f1*100:0.2f}%"
    ])
performance_chart = pd.DataFrame(performance_data,columns=["Model","Test Accuracy","Precision", "Recall", "F1"])
performance_chart
```

### Best Model

```python
best_model = decision_model()
```
